```python
Dictionary = {'a':0,
              'b':1,
              'c':2,
              'd':3,
              'e':4,
              'f':5,
              'g':6,
              'h':7,
              'i':8,
              'j':9,
              'k':10,
              'l':11,
              'm':12,
              'n':13,
              'o':14,
              'p':15,
              'q':16,
              'r':17,
              's':18,
              't':19,
              'u':20,
              'v':21,
              'w':22,
              'x':23,
              'y':24,
              'z':25,
              ' ':26,
              '-':27}
reverse = ['a',
              'b',
              'c',
              'd',
              'e',
              'f',
              'g',
              'h',
              'i',
              'j',
              'k',
              'l',
              'm',
              'n',
              'o',
              'p',
              'q',
              'r',
              's',
              't',
              'u',
              'v',
              'w',
              'x',
              'y',
              'z',
              ' ',
              '-']

x = 'live long and prosper'
K = [[1,1,2],[4,1,4],[1,3,5]]
def function(v1):
    res = [0,0,0]
    res[0] = K[0][0]*v1[0]+K[0][1]*v1[1]+K[0][2]*v1[2]
    res[1] = K[1][0]*v1[0]+K[1][1]*v1[1]+K[1][2]*v1[2]
    res[2] = K[2][0]*v1[0]+K[2][1]*v1[1]+K[2][2]*v1[2]
    res[0] = res[0]%28
    res[1] = res[1]%28
    res[2] = res[2]%28
    return res

y = ''   
print(K)
isDone = False
i = 0
while not isDone:
    tmp = x[i:i+3]
    print(tmp)
    i = i+3
    if i > len(x):
        isDone = True
        break
    vektor = []
    vektor.append(Dictionary[tmp[0]])
    vektor.append(Dictionary[tmp[1]])
    vektor.append(Dictionary[tmp[2]])
    index = function(vektor)
    y = y + (reverse[index[0]])
    y = y + (reverse[index[1]])
    y = y + (reverse[index[2]])
print(y)
```
